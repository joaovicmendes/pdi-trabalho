# Leitura e plotagem de imagens utilizando Python

```python
import matplotlib.pyplot as plt   # Biblioteca utilizada para plotar figuras
                                  # Pode ser usada tamb√©m para ler figuras do disco
```

```python
# Leitura da imagem utilizando o matplotlib
img = plt.imread('earth_gray.jpg')
```

```python
# Plotagem da imagem
plt.imshow(img)
```

```python
# Plotagem da imagem usando o mapa de cores gray
plt.imshow(img, cmap='gray')
```

```python
# Leitura da imagem colorida
img = plt.imread('earth.jpg')
plt.imshow(img)
```

```python
# Mostra cada canal de cor da imagem separadamente
img_red = img[:,:,0]
img_green = img[:,:,1]
img_blue = img[:,:,2]

plt.figure(figsize=[15, 5.])
plt.subplot(1, 3, 1)
plt.imshow(img_red, cmap='gray')
plt.title('Canal vermelho')
plt.subplot(1, 3, 2)
plt.imshow(img_green, cmap='gray')
plt.title('Canal verde')
plt.subplot(1, 3, 3)
plt.imshow(img_blue, cmap='gray')
plt.title('Canal azul')
plt.tight_layout()
```
